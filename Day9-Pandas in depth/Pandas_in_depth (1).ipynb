#### Drop 
---

Deleting a column 
- axis 1 = Column
- axis 0 = Row

```python
newdf.drop(0, axis=1)
```

```python
newdf.loc[[1,2], ['C','D']]
```

*To get all the column or row values use ':'*

```python
newdf.loc[:, ['C','D']]
```

```python
newdf.loc[[1,2],:]
```

### To find in which all rows values are greater than a particular another value

```python
newdf.loc[(newdf['A']<0.3) & (newdf['C']>0.1)]
```
