# Matplotlib
---
- plt.plot(x_name, y_name) : plotting the given y & x axis.
- plt.title() : for giving a title to the plot/graph
- plt.xlabel("year") : labeling the x axis 
- plt.ylabel("Income in USD") : labeling the y axis
- plt.yticks(income_ticks, [f"${x}k" for x in income_ticks]) : thos is for correct labeling the values on y axis.
- plt.show() : showing the plotted graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Customization

```python
years = [2014, 2015, 2016, 2017,
        2018, 2019, 2020, 2021]

income = [55, 56, 62, 61,
         72, 72, 73, 75]
```

```python
income_ticks = list(range(50, 81,2))
```

```python
plt.plot(years, income)
plt.title("Income of John(in USD)" , fontsize = 15, fontname='Serif')
plt.xlabel("Year", fontsize = 10, fontname='Serif')
plt.ylabel("Income in USD", fontsize = 10, fontname='Serif')
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```

## Subplotting
---

```python
x = np.arange(100)
```

```python
fig, axs = plt.subplots(2, 2) 

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log Function")
axs[1, 1].set_xlabel("TEST")

fig.suptitle("Four Plots")
#plt.savefig("fourplots.png", dpi=300, transparent=False, pad_inches=20.2)

plt.show()
```

## 3D Plotting
---

```python
ax = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

### Surface 3D Plots
---

```python
ax = plt.axes(projection="3d")

x= np.arange(-5, 5, 0.1)
y= np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap="Spectral")    #cmap is coloring parameter
ax.set_title("3D Plot")
ax.set_xlabel("test")
#ax.view_init(azim = 0, elev=90)       #for view from above
plt.show()
```

## Animating Plots
---

```python
import random
```

```python
heads_tails = [0,0]

for _ in range(100000):
    heads_tails[random.randint(0,1)] += 1
    plt.bar(["Heads", "Tails"], heads_tails, color=["red", "blue"])
    plt.pause(0.001)

plt.show()
```
