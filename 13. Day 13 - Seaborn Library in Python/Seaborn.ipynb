# Seaborn 

```python
import seaborn as sns
```

## Loading the Seaborn Dataset

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

## Seaborn Scatter Plot
---
- palette - lets one decide the color of the plots

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")
```

## Histograms and Distribution Plots

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

## Bar chart

```python
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

## Box plot

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

  ## Strip plot
  ---

  'dodge' hepls in segregating mixed 2 diff kinds of values in the plot. Like here, female and male genders are separated

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True, palette="YlGnBu")
```

## Joint plot

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind='reg')
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind='kde', fill=True, cmap="YlGnBu")   #kde plot
```

Pair Plot

```python
sns.pairplot(titanic.select_dtypes(['number']), hue="pclass")
```

## Heat Map

```python
import numpy as np 
import seaborn as sn 
import matplotlib.pyplot as plt 
  
data = np.random.randint(low = 1, 
                         high = 100, 
                         size = (10, 10)) 
print("The data to be plotted:\n") 
print(data) 
  

hm = sn.heatmap(data = data, cmap="YlGnBu") 
plt.show()
```
