# Matplot library 

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plot

---
#### plt.scatter
- c = color
- marker = type of marking you need
- alpha = used to show merging
- s = size of graph

```python
X_data =np.random.random(500) * 100
y_data =np.random.random(500) * 100
```

```python
plt.scatter(X_data, y_data, c='red', s=150, marker="*", alpha=0.3)
plt.show()
```

## Line Plots

---
#### plt.plot
- lw = thickness of the line 
- linestyle = styling the lne
- r-- = red colour with -- line style

```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85,86, 82, 81, 79, 83, 80, 82, 82, 83, 81, 80, 79]
```

```python
plt.plot(years, weights, "r--", lw=3)    
plt.show()
```

## Bar plots

---
#### plt.bar
- align=edge (this givs the bar to start from the very rigth point of the given values on the x axis)
- here color is given by color='b'
- width = width of each bars
- edgecolor = color of the edges

```python
x = ["c++", "C#", "Python", "Java", "Go"]
y = [20, 45, 56, 83, 45]
```

```python
plt.bar(x, y, color="b", align="edge", width=0.5, edgecolor="green", lw=3)
plt.show()
```

## Histogram

---
#### plt.hist
- bins = more no.of bars

```python
ages = np.random.normal(20,1.5,1000)
```

```python
plt.hist(ages, bins=[ages.min(), 18, 21, ages.max()])    #here, max bin will till 21 and min till 18.
plt.show()
```

## Pie charts

---
#### plt.pie

- explode = removes apart a particular section from the pie chart.
- autopct = automatically gives the percentage for each section.
- pctdistance = enables to format and make a distance the given percentages in the section accordingly.
- startangle = adjust the pie chart according to our view.

```python
langs = ["C++", "C#", "Python", "Java", "Go"]
votes = [50, 24, 14, 6, 17]
explodes = [0, 0, 0, 0.8, 0]
```

```python
plt.pie(votes, labels=langs, explode=explodes, 
       autopct="%.2f%%", pctdistance=0.7, startangle=90)
plt.show()
```
